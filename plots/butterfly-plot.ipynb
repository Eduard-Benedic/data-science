```python
import numpy as np
import matplotlib.pyplot as plt


fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
fig.set_size_inches(35, 35)
fig.set_dpi(120)

def func(x, y):
    return x**2*np.sin(y)

X = np.linspace(-22, 22, 100)
Y = np.linspace(-1, 1,100)
X, Y = np.meshgrid(X, Y)

z = func(X, Y)

ax.set_title('Graph function for x^2 * sin(y)', fontdict={'fontsize': 40})
xAxis = ax.get_xaxis()
yAxis = ax.get_yaxis()
xAxis.set_label_text('X axis', fontdict={'fontsize': 40, 'color': 'green'})
yAxis.set_label_text('Y axis', fontdict={'fontsize': 40, 'color': 'green'})
ax.set_zlabel('Z axis')
ax.view_init(-5, 280)
ax.plot_surface(X, Y, z)

fig.savefig('3D plot surface')
```
