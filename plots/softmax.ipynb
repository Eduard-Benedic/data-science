```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib notebook

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
fig.set_size_inches(35, 35)
fig.set_dpi(120)

def softmax(x, y):
  return np.e**x / (np.e**x + np.e**y)

x = np.linspace(-100, 100, 50)
y = np.linspace(-100, 100, 50)

X, Y = np.meshgrid(x, y)
Z = softmax(X, Y)
ax.view_init(0, 90)


ax.set_title('Graph function for softmax', fontdict={'fontsize': 40})
xAxis = ax.get_xaxis()
yAxis = ax.get_yaxis()
xAxis.set_label_text('X axis', fontdict={'fontsize': 20, 'color': 'green'})
yAxis.set_label_text('Y axis', fontdict={'fontsize': 20, 'color': 'green'})
ax.view_init(170,  380)
ax.plot_surface(X, Y, Z)

fig.savefig('Softmax plot surface')
plt.show()
```
