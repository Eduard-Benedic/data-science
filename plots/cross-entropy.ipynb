```python
import numpy as np
import matplotlib.pyplot as plt

def cross_entropy(x):
  return -np.log(x)
x = np.linspace(0.01, 1, 100)
y = cross_entropy(x)

fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
fig.set_dpi(120)

ax.set_xlabel('x label')
ax.set_ylabel('y label', { 'fontsize': 40 })
ax.plot(x, y)

fig.savefig('Cross entropy')

# print(y)
```
